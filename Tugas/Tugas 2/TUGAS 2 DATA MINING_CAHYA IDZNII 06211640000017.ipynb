## Missing Value

```python
data.info()  #informasi data jumlah data tiap variabel jika kurang dari 303 maka terdapat missing value
```

```python
np.sum(data.isnull()) #jumlah missing value tiap variabelnya
```

```python
data.isnull().sum().sum() #jumlah semua missing value pada data
```

```python
#imputasi missing value data numerik dengan nilai mean (metode listwise)
data['trestbps'] = data['trestbps'].fillna((data['trestbps'].mean())) 
data['chol'] = data['chol'].fillna((data['chol'].mean()))
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk age,sex,cp,chol,fbs,dan target tanpa membuat kolom baru
data['age'] = LabelEncoder().fit_transform(data['age'])
data['sex'] = LabelEncoder().fit_transform(data['sex'])
data['cp'] = LabelEncoder().fit_transform(data['cp'])
data['chol'] = LabelEncoder().fit_transform(data['chol'])
data['fbs'] = LabelEncoder().fit_transform(data['fbs'])
data['target'] = LabelEncoder().fit_transform(data['target'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Standardize untuk data
data['age'] = StandardScaler().fit(data[['age']]).transform(data[['age']])
data['trestbps'] = StandardScaler().fit(data[['trestbps']]).transform(data[['trestbps']])
data['chol'] = StandardScaler().fit(data[['chol']]).transform(data[['chol']])
data['thalach'] = StandardScaler().fit(data[['thalach']]).transform(data[['thalach']])
data['target'] = StandardScaler().fit(data[['target']]).transform(data[['target']])
data.head()
```

## Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
#Normalize untuk data
data['age'] = Normalizer().fit(data[['age']]).transform(data[['age']])
data['trestbps'] = Normalizer().fit(data[['trestbps']]).transform(data[['trestbps']])
data['chol'] = Normalizer().fit(data[['chol']]).transform(data[['chol']])
data['thalach'] = Normalizer().fit(data[['thalach']]).transform(data[['thalach']])
data['target'] = Normalizer().fit(data[['target']]).transform(data[['target']])
data.head()
```
