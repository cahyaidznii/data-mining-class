```python
import pandas as pd
import numpy as np
```

```python
# Menggabungkan DataFrame
Data = pd.DataFrame({'No': ['No0', 'No1', 'No2', 'No3'],
                     'JK': ['P', 'P', 'L', 'L'], # JK=Jenis Kelamin
                     'Usia': ['15', '17', '18', '20']})

Grade = pd.DataFrame({'No': ['No0', 'No1', 'No2', 'No3'],
                      'IPK': ['3.78', '3.45', '3.65', '3.00'],
                      'Nilai': ['A', 'B', 'AB', 'B']})

Hasil1 = pd.merge(Data, Grade, on='No')
Hasil1
```

```python
# Mengurutkan Tabel
Hasil1 = Hasil1[['No', 'JK', 'Usia', 'Nilai','IPK']]
Hasil1
```

```python
# Memanggil data dengan multiple key
Data = pd.DataFrame({'No': ['No0', 'No0', 'No1', 'No1'],
                        'Kelas': ['Statmat', 'AED', 'Datmin', 'ADW'],
                        'JK': ['P', 'P', 'L', 'L'], 
                      'Usia': ['15', '17', '18', '20']})

Grade = pd.DataFrame({'No': ['No0', 'No0', 'No1', 'No1'],
                      'Kelas': ['Statmat', 'ADK', 'MO', 'Datmin'],
                      'IPK': ['3.78', '3.45', '3.65', '3.00'],
                      'Nilai': ['A', 'B', 'AB', 'B']})

Hasil2 = pd.merge(Data, Grade, on=['No', 'Kelas'])
Hasil2
```

```python
# Metode penggabungan (default how="inner")
result_left = pd.merge(Data, Grade, how='left', on=['No', 'Kelas'])
result_left
```

```python
# Metode penggabungan (default how="inner")
result_right = pd.merge(Data, Grade, how='right', on=['No', 'Kelas'])
result_right
```

```python
result_outer = pd.merge(Data, Grade, how='outer', on=['No', 'Kelas'])
result_outer
```

```python
result_inner = pd.merge(Data, Grade, how='inner', on=['No', 'Kelas'])
result_inner
```

```python
## Joining 
Data = pd.DataFrame({'JK': ['P', 'P', 'L'], 
                      'Usia': ['18', '20', '18']},
                    index=['No0', 'No1', 'No2'])
Grade = pd.DataFrame({'Kelas': ['Statmat', 'ADK', 'Datmin'],
                      'IPK': ['3.85', '3.45', '3.00']},
                    index=['No0', 'No2', 'No3'])
result = Data.join(Grade) #Based Data ditambahkan Grade
result
```

```python
#Seluruh data masuk
result = Grade.join(Data, how='outer') 
result
```

```python
#hanya data lengkap
result = Grade.join(Data, how='inner') 
result
```

```python
# Penyelarasan data dalam indeks (label baris).
#Perilaku yang sama ini dapat dicapai menggunakan gabungan ditambah argumen tambahan yang memerintahkannya untuk menggunakan indeks
result = pd.merge(Data, Grade, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(Data, Grade, left_index=True, right_index=True, how='inner')
result
```

```python
Data = pd.DataFrame({'JK': ['P', 'L', 'P', 'L'], 
                      'Usia': ['19', '21', '18', '23'],
                   'key': ['No0', 'No1', 'No0', 'No1']})
Grade = pd.DataFrame({'Grade': ['A', 'B'],
                       'IPK': ['3.44', '3.74']},
                      index=['No0', 'No1'])
result = Data.join(Grade, on='key')
result
```

```python
d1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
d2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
d3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [d1, d2, d3]
result = pd.concat(frames)
result
```

```python
# menambah kategori
result = pd.concat(frames, keys=['X', 'Y', 'Z'])
result
```

```python
# join semua row
d4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([d1, d4], axis=0) 
result
```

```python
# Join = inner (gabungan dengan baris yang ada dikedua data)
# hanya yang ada di d1 dan d4
result = pd.concat([d1, d4], axis=1, join='inner')
result
```

```python
#menggabungkan
result = d1.append(d2) #default axis=0
result
```

```python
#mengabungkan banyak dataset
result = d1.append([d2, d3]) 
result
```

```python
#menambah baris ke data 
s2 = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
result = d1.append(s2, ignore_index=True)
result
```
