```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#IMPORT DATA DARI CSV
datauts =pd.read_csv('online_retail.csv', sep=',')
datauts
```

```python
# DATA CLEANING
datauts['Description'] = datauts['Description'].str.strip()
datauts.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
  
datauts['InvoiceNo'] = datauts['InvoiceNo'].astype('str') 
datauts = datauts[~datauts['InvoiceNo'].str.contains('C')]
```

```python

hasil = (datauts.groupby(['InvoiceNo', 'Description'])['Quantity']           
.sum().unstack().reset_index().fillna(0)           
.set_index('InvoiceNo'))
hasil.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
hasil_sets = hasil.applymap(encode_units)
 
#menghapus kolom postage
hasil_sets.drop('POSTAGE', inplace=True, axis=1)
hasil_sets.head()
```

```python
# Build up the frequent items 
# menentukan minimal support dengan minimal 0,025 atau 2,5%
frequent_itemsets = apriori(hasil_sets, min_support=0.025, use_colnames=True)
frequent_itemsets.head()
```

```python
# Create the rules ( MELAKUKAN PROSESES ASSOSIATION RULE)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# menentukan nulai lift minimal 6 dan confident minimal 0.2
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.2) ]
```
